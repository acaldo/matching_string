```python
import itertools
import pandas as pd
# The new library!
from thefuzz import fuzz, process
```

```python
from data import array1, array2
```

```python
df1 = pd.DataFrame(
    array1,
    columns =['CLIENT']
)
df2 = pd.DataFrame(
    array2,
    columns =['Firm Name']
)
```

```python
df1
```

#### Data Preprocessing
1. Create the df dataframe containing the product of the two CSVs

```python
list(itertools.product(df1['CLIENT'].values,df2['Firm Name'].values))
```

```python
df = pd.DataFrame(
    list(itertools.product(df1['CLIENT'].values,df2['Firm Name'].values)),
    columns =['CSV 1', 'CSV 2']
)
df.shape
```

2. Create a new column Ratio Score that contains the distance for all the rows in df
score

```python
score = [fuzz.partial_ratio(c1,c2) for c1, c2 in df.values]
##Ratio Score
df['Ratio Score'] = score
df.head()
```

3. How many rows have a Ratio score of 90 or more?

```python
df[df['Ratio Score'] >= 90]['Ratio Score'].count()

## with query
# df.query("`Ratio Score` >= 90")
```

4. What's the corresponding company in CSV2 to AECOM in CSV1?

```python
df[(df['CSV 1'] == 'AECOM') & (df['Ratio Score'] > 90) ]['CSV 2']

#df.query("`CSV 1` == 'AECOM' and `Ratio Score` > 90")
```

5. What's the corresponding CSV2 company of Starbucks?

```python
df[(df['CSV 1'] == 'Starbucks') & (df['Ratio Score'] > 90) ]['CSV 2']
```

6. Is there a matching company for Pinnacle West Capital Corporation?


```python
df[df['CSV 1'] == 'Pinnacle West Capital Corporation'].sort_values(by='Ratio Score', ascending= False)
```

7. How many matching companies are there for County of Los Angeles Deferred Compensation Program?


```python
df[df['CSV 1'] == 'County of Los Angeles Deferred Compensation Program'].sort_values(by='Ratio Score', ascending= False)
```
